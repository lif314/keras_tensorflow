```python
import keras
import numpy as np
import matplotlib.pyplot as plt
#Sequential 按顺序构成的模型
from keras.models import Sequential
# Dense全连接层
from keras.layers import Dense
```

```python
# 使用numpy生成100个随机数据
x_data = np.random.rand(100)
# 噪音，随机点
noise = np.random.normal(0,0.01, x_data.shape) # 噪声
y_data = x_data*0.1 + 0.2 + noise   # 线性模型

# 显示随机点
plt.scatter(x_data, y_data) # 散点图
plt.show()
```

```python
# 构建一个顺序模型
model = Sequential()
 # Shit + 双击Tab键查看参数说明
# 在模型中添加一个全连接层
model.add(Dense(units=1, input_dim=1))  # units输出维度  input_dim输入维度
# 编译模型
model.compile(optimizer='sgd', loss='mse') # 随机梯度下降  均方误差
# 训练: 一般采用batch训练，每次放一些批次
for step in range(3000):
    cost = model.train_on_batch(x_data, y_data)  # 批次训练
    if step % 500 == 0:
        print("cost:", cost)
# 打印权值和偏置值
W, b = model.layers[0].get_weights()
print('W:', W, ',b:', b)

# 将x_data放在模型中进行预测
y_pred = model.predict(x_data)
# 显示随机点
plt.scatter(x_data, y_data)
# 显示预测结果
plt.plot(x_data, y_pred, 'r-', lw=3)  # 预测的一条直线
plt.show()
```

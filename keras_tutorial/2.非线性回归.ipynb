# <center> 2.非线性回归 </center>

```python
import keras
import numpy as np
import matplotlib.pyplot as plt
#Sequential 按顺序构成的模型
from keras.models import Sequential
# Dense全连接层
from keras.layers import Dense, Activation # 激活函数
from keras.optimizers import SGD # 随机梯度下降
```

```python
# 生成数据
x_data = np.linspace(-0.5, 0.5, 200) # -0.5到0.5的200个均匀分布的点
noise = np.random.normal(0, 0.02, x_data.shape)  # 噪声
y_data = np.square(x_data) + noise  # 对点取平方

# 显示随机点
plt.scatter(x_data, y_data)
plt.show()
```

```python
# 构建一个顺序模型 --- 线性回归
model = Sequential()
 # Shit + 双击Tab键查看参数说明
# 在模型中添加一个全连接层
model.add(Dense(units=1, input_dim=1))  # units输出维度  input_dim输入维度
# 编译模型
model.compile(optimizer='sgd', loss='mse') # 随机梯度下降  均方误差
# 训练: 一般采用batch训练，每次放一些批次
for step in range(3000):
    cost = model.train_on_batch(x_data, y_data)  # 批次训练
    if step % 500 == 0:
        print("cost:", cost)
# 打印权值和偏置值
W, b = model.layers[0].get_weights()
print('W:', W, ',b:', b)

# 将x_data放在模型中进行预测
y_pred = model.predict(x_data)
# 显示随机点
plt.scatter(x_data, y_data)
# 显示预测结果
plt.plot(x_data, y_pred, 'r-', lw=3)  # 预测的一条直线
plt.show()
```

```python
# 构建一个非线性回归模型
model = Sequential()
# 在模型中添加一个全连接层
# 多添加隐藏层(10个神经元) 1-10-1 
# 默认是线性的激活函数  非线性需要添加激活函数
# 可以自定义添加激活函数，或者在参数列表中指定
model.add(Dense(units=10, input_dim=1, activation='relu'))  # 添加一个隐藏层，输出10个维度
# model.add(Activation('tanh')) # 添加激活函数 注释 ctrl + ?
model.add(Dense(units=1)) # 输出层，不需要输入层数作为参数，可以根据上一层知道是10
model.add(Activation('tanh')) # 添加激活函数

# 输入层(1)-隐藏层(10)-激活函数-输出层-激活函数 

# 编译模型
# sgd: Stochastic gradient descent 随机梯度下降【默认】学习率较小，迭代次数多
# mse: Mean Squared Error 均方误差
# 使用自定义的SGD
sgd = SGD(learning_rate=0.3) # 学习率
model.compile(optimizer=sgd, loss='mse')   
# 训练: 一般采用batch训练，每次放一些批次
for step in range(3000):
    cost = model.train_on_batch(x_data, y_data)  # 批次训练
    if step % 500 == 0:
        print("cost:", cost)
# 打印权值和偏置值
W, b = model.layers[0].get_weights()
print('W:', W, ',b:', b)

# 将x_data放在模型中进行预测
y_pred = model.predict(x_data)
# 显示随机点
plt.scatter(x_data, y_data)
# 显示预测结果
plt.plot(x_data, y_pred, 'r-', lw=3)  # 预测的一条直线
plt.show()
```
